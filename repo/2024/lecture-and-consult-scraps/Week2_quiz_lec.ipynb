```julia
"""
The function takes a string `s` and returns the first index where the character `c`
appears, or returns `0` if there character is not in the string.
"""
function find_first_char_index(s::String, c::Char)
    for (i,current_char) ∈ enumerate(s)
        if current_char == c
            return i
        end
    end
    return 0
end
```

```julia
find_first_char_index("hello eye",'b')
```

```julia
find_first_char_index("hello eye",'e')
```

```julia
typeof(find_first_char_index("hello",'c'))
```

```julia
? find_first_char_index
```

```julia
function find_first_char_rec(s, c, i)
    if length(s) == 0
        return 0
    end
    if s[1] == c
        return i
    end
    return find_first_char_rec(s[2:end], c, i+1)
end
```

```julia
find_first_char_rec("hello world", 'e', 1)
```

```julia
find_first_char_rec("hello world", 'q', 1)
```

```julia
find_first_char_rec("eello world", 'e', 1)
```
