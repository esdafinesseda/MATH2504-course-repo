```julia
X = df[:,[:GREScore, :TOEFLScore, :CGPA, :UniversityRating ]] |> Tables.Matrix |> float
```

```julia
heatmap(X, yflip=true, legend = false)
```

```julia
n, d = size(df)
```

```julia
using Random
Random.seed!(0)
p_validate = 0.2
n_validate = floor(Int, p_validate*n)
validate_index_set = sample(1:n, n_validate, replace = false)
train_index_set = setdiff(1:n, validate_index_set);
```

```julia
n_validate = length(validate_index_set)
n_train = length(train_index_set)
n_train, n_validate
```

```julia
y = df.ChanceofAdmit;
histogram(y,legend=false,bins=30)
```

### Decision Trees

```julia
using DecisionTree
```

```julia
y_train, X_train = y[train_index_set], X[train_index_set, :];
y_validate, X_validate = y[validate_index_set], X[validate_index_set, :];
```

[DecisionTree.jl](https://github.com/bensadeghi/DecisionTree.jl)

```julia
tree_model = build_tree(y_train, X_train, 4, 5) #See DecisionTree.jl for features
```

```julia
[apply_tree(tree_model,X_train[k,:]) for k in 1:n_train]
```

```julia
apply_tree(tree_model,X_train)
```

```julia
y_hat_validate = apply_tree(tree_model,X_validate)
```

```julia
scatter(y_validate, y_hat_validate, 
        label = false, 
        xlabel = "Actual Data", 
        ylabel = "Predicted Data")
```

### Random Forests

```julia
forest_model = build_forest(y_train, X_train)
```

```julia
y_hat_validate_forest = apply_forest(forest_model, X_validate)
```

```julia
scatter(y_validate, [y_hat_validate y_hat_validate_forest], 
        label = ["Single tree" "Forest"], 
        xlabel = "Actual Data", 
        ylabel = "Predicted Data", legend = :topleft)
```
